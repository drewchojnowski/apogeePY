## apogeespec.py also has a continuum normalization routine, but unfortunately it can't be run within this ipython notebook. The syntax would be as follows:

(1) bla=x.apcontinuum('../bep_spectra/apVisit-apogee2-57194-2MASS-J18273952-0349520.fits')

(2) a window will pop up showing the blue detector spectrum

(3) spline fitting points can be selected either with left click on the mouse (right click to delete point) to mark a point on the continuum, or 'z' key to mark a point above or below the continuum.

(4) pressing 'enter' displays the fit on the plot

(5) pressing 'n' applies the fit to the spectrum and shows the result

(6) this will be repeated for the other two detectors.

(7) finally, a file will be written out with "_norm" appended to the file name.

## There is also a plotting routine within apogeespec.py called "apsplot". This is intended to be similar to "SPLOT" in IRAF.

```python
a=x.apsplot('../bep_spectra/apVisit-apogee2-57194-2MASS-J18273952-0349520.fits')
```
